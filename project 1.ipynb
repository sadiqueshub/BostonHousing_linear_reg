```python
### separating Target and independent variable
x = boston_df.iloc[:,:-1]
y = boston_df.iloc[:,-1].to_frame()
```

#### train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 42)
y_train
```

### standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)

x_test = scaler.transform(x_test)
x_train
```

### Model Training


```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train,y_train)
regression.coef_
```

```python
## getting coefficients_ and intercepts_
print(regression.coef_)
```

```python
print(regression.intercept_)
```

### Prediction Time

```python
pred = regression.predict(x_test).flatten()
pred
```

```python
### scatter plot to see the accuracy of model

plt.scatter(y_test,pred)
plt.show()
```

```python
residuals = pred - y_test['Price']
residuals
sns.displot(residuals,kind = "kde")
plt.show()
```

### Accuracy Metrics

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,pred))
print(mean_absolute_error(y_test,pred))
print(np.sqrt(mean_squared_error(y_test,pred)))
```

### R square and Adjusted R square

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,pred))
```
